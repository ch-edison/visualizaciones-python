```python
data2 = []
lista02 = list(data02.CANDIDATO_NOMBRE)
lista04 = list(data04.CANDIDATO_NOMBRE)
lista06 = list(data06.CANDIDATO_NOMBRE)
lista07 = list(data07.CANDIDATO_NOMBRE)
lista09 = list(data09.CANDIDATO_NOMBRE)
lista13 = list(data13.CANDIDATO_NOMBRE)
lista14 = list(data14.CANDIDATO_NOMBRE)

lista_total = []
lista_total.extend(lista02)
lista_total.extend(lista04)
lista_total.extend(lista06)
lista_total.extend(lista07)
lista_total.extend(lista09)
lista_total.extend(lista13)
lista_total.extend(lista14)
lista_total = list(set(lista_total))
for l in lista_total:
    diccionario = {}
    diccionario['nombre'] = l.strip()
    if l in lista02:
        diccionario['2002'] = 1
    if l in lista04:
        diccionario['2004'] = 1
    if l in lista06:
        diccionario['2006'] = 1
    if l in lista07:
        diccionario['2007'] = 1
    if l in lista09:
        diccionario['2009'] = 1
    if l in lista13:
        diccionario['2013'] = 1
    if l in lista14:
        diccionario['2014'] = 1
     
    data2.append(diccionario)

# data2 = [{'nombre': 1, '2002': 2}, {'nombre': 5, '2004': 10, '2002': 20}]

data = pd.DataFrame(data2)
print data.keys()
```
