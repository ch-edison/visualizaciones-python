```python
meses = []
ocurrencia = []
for d in data[data['edad']<15].edad.unique():
    meses.append(d)
    ocurrencia.append(data[data['edad']==d]['sector'].count())
```

```python
print meses
print ocurrencia
```

```python
from bokeh.charts import Area, show, vplot, output_file

# create some example data
data2 = dict(
    meses=meses,
    ocurrencia=ocurrencia)

area = Area(data2, title="Area Chart", legend="top_left",
            xlabel='día', ylabel='ocurrencia')
bokeh.io.show(area)
```

```python
provincias = data.prov_res.unique()

provs = []
ocurrencia = []
for d in provincias:
    provs.append(d)
    ocurrencia.append(data[data['prov_res']==d]['cod_est'].count())

print provs
print ocurrencia
```

```python
from bokeh.charts import Bar, output_file, show, hplot

datos1 = {
    'interpreter': provincias,
    'timing': ocurrencia
}
bar = Bar(datos1, values='timing', label='interpreter', legend=None,
          title="Ocurrencias por provincia", width=800)
bokeh.io.show(bar)
```

```python
from bokeh.charts import Bar, output_file, show, hplot
dias = data.dia_ingr.unique()
print dias

for mes in ['Enero', 'Febrero', 'Marzo']:
    l_dias = []
    ocurrencia = []
    for d in dias:
        l_dias.append(d)
        # datos_tmp = 
        ocurrencia.append(data[(data['mes_ingr']==mes) & (data['dia_ingr']==d)]['sector'].count())

    print l_dias
    print ocurrencia


    data2 = {
        'interpreter': l_dias,
        'timing': ocurrencia
    }
    bar = Bar(data2, values='timing', label='interpreter', legend=None,
          title="Ocurrencias por meses de %s" % mes, width=800)
    bokeh.io.show(bar)
```
